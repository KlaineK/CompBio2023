```python
from Bio.PDB.PDBParser import PDBParser
parser = PDBParser()
```

```python
#get_structure(id, target)
structure_pdb = parser.get_structure('6lu7', './6lu7.pdb')
structure_pdb 
```

```python
#dalam 1 file ada banyak struktur.
# tapi dalam file ini hanya bisa menampung 1 struktur 
model = structure_pdb[0]
model

#model = structure_pdb[1] akan error.
```

```python
#mengambil data dari model. mencari chain
for chain in model:
    print(chain)
    #mencari residue
    for residue in chain:
        print(residue)
        #mencari atom yang membentuk residue
        for atom in residue:
            print(atom)
```

```python
#FASTA file
from Bio import SeqIO
dir(SeqIO)
```

```python
#parse(handle, format)
#handle : jenis file yang akan dihandle. berisi path file
#format : format file yang akan dihandle 
fasta_record = SeqIO.parse('./coronavirus.fasta', 'fasta')
for record in fasta_record:
    print(record.id)
    print(record.description)
    print(record.seq)
    for protein in record.seq.translate().split('*'):
        print (protein)
        print (len(protein))
```
