```python
# find codon position
seq2 = Seq('CGATCGATCGATCACA')
print("sequence length")
print(len(seq2))
print(seq2.find('A'))

print(seq2.find('TCG'))

#find codon position from right 
print(seq2.rfind('C'))

#cant find codon
print(seq2.find('o'))

#find codon index position
print('Index search')
print(seq2.index('A'))
print(seq2.rindex('G'))

#cant find codon index
#resulting error
print(seq2.index('p'))
```

```python
#import pyplot from matplotlib
from matplotlib import pyplot as plt

#import counter from collections
from collections import Counter
```

```python
mainsq = Seq('AGCTAGCTAGCGTGC')
#count total codons in a sequence
dna_freq = Counter(mainsq)
print(dna_freq)

#look variance of codons
#displaying the key components
print(dna_freq.keys())

#look total counter of codons
#displaying the frequency values of key components
print(dna_freq.values())
```

```python
#creating bar chart
#struktur : plt.bar(x-axis, y-axis)
plt.bar(dna_freq.keys(), dna_freq.values())
```

```python
mainsq2 = Seq('CAGTCGATGC')
mainsq2
```
